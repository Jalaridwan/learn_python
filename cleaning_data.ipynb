### Melihat dan merubah tipe data

```python
ride_sharing['duration'] = ride_sharing['duration'].str.strip('minutes')
ride_sharing['duration'] = ride_sharing['duration_time'].astype('int')
```

```python
ride_sharing.head()
```

```python
# menghapus kolom yang tidak dibutuhkan
ride_sharing.drop('Unnamed: 0', 1, inplace = True)
```

```python
ride_sharing.head()
```

```python
# menegaskan tipe data sudah berubah
assert ride_sharing['duration'].dtype == 'int'
ride_sharing.info()
```

```python
## cek user type

ride_sharing['user_type'].unique()
```

```python
ride_sharing['user_type'] = ride_sharing['user_type'].astype('category')
```

```python
assert ride_sharing['user_type'].dtype == 'category'
```

```python
ride_sharing['user_type'].describe()
```

## MEMANIPULASI DATA CATEGORY

Seringkali terdapat beberapa data kategori yang tidak sesuai, jadi kita harus memanipulasinya.

### menghapus data selain dalam kategori

```python
#kategori

category = pd.DataFrame({'kategori':['makanan', 'minuman']})
category
```

```python
# data
df_warkop = pd.DataFrame({'Barang':['kopi','mie','teh', 'gorengan', 'rokok', 'bubur'], 'kategori':['minuman','makanan','minuman','makanan','lain-lain','makanan']})
df_warkop
```

```python
beda_kategori = set(df_warkop['kategori']).difference(category['kategori'])
beda_kategori
```

```python
df_warkop_selain_kategori = df_warkop['kategori'].isin(beda_kategori)
df_warkop_selain_kategori
```

```python
df_warkop[~df_warkop_selain_kategori]
```
